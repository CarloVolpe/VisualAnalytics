```python
import os
import pandas as pd
import numpy as np
```

```python
folder_path = 'C:/Users/Enrico/Desktop/Datasets/Activity Logs'

dfs = []

for file_name in os.listdir(folder_path):
    if file_name.endswith(".csv"):
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path)
        dfs.append(df)
```

```python
merged_df = pd.concat(dfs, ignore_index=True)
merged_df.info()
```

```python
activity_logs_df = merged_df
activity_logs_df.info()
```

```python
# Conversione della colonna 'timestamp' in formato datetime
activity_logs_df['timestamp'] = pd.to_datetime(activity_logs_df['timestamp'])

# Estrazione della parte della data (anno-mese-giorno)
activity_logs_df['date'] = activity_logs_df['timestamp'].dt.strftime('%Y-%m-%d')
```

```python
activity_logs_df = activity_logs_df[activity_logs_df["date"]== "2022-03-01"]
```

```python
activity_logs_df
```

```python
travel_journal_df = pd.read_csv("C:/Users/Enrico/Desktop/Datasets/Journals/TravelJournal.csv")
```

```python
travel_journal_df.info()
```

```python
travel_journal_df['travelStartTime'] = pd.to_datetime(travel_journal_df["travelStartTime"])


travel_journal_df['date'] = travel_journal_df['travelStartTime'].dt.strftime('%Y-%m-%d')
```

```python
travel_journal_df = travel_journal_df[travel_journal_df["date"]== "2022-03-01"]
```

```python
travel_journal_df = travel_journal_df[travel_journal_df["purpose"]=='Work/Home Commute']
```

```python
travel_journal_df
```

```python


# Inizializzare una lista per memorizzare i risultati
result_rows = []

# Iterare attraverso ogni riga del travel_journal
for index, row in travel_journal_df.iterrows():
    # Estrarre l'intervallo di tempo e il participantId
    start_time = pd.Timestamp(row['travelStartTime'])
    end_time = pd.Timestamp(row['travelEndTime'])
    participant_id = row['participantId']
    
    # Filtrare le righe del activity_logs per participantId e intervallo di tempo
    filtered_logs = activity_logs_df[(activity_logs_df['participantId'] == participant_id) &
                                  (activity_logs_df['timestamp'] >= start_time) &
                                  (activity_logs_df['timestamp'] <= end_time)]
    
    # Creare una lista dei currentLocation corrispondenti
    current_locations = filtered_logs['currentLocation'].tolist()
    
    # Aggiungere le informazioni al risultato
    result_rows.append({
        'participantId': participant_id,
        'intervalStartTime': start_time,
        'intervalEndTime': end_time,
        'currentLocations': current_locations
    })

# Costruire il dataframe dei risultati
result_df = pd.DataFrame(result_rows)

result_df
```

```python
result_df_sorted = result_df.sort_values(by="participantId")
```

```python
result_df_sorted.to_csv("Commute_activity.csv", index=False)
```
