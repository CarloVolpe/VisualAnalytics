```python
import os
import pandas as pd
import numpy as np
```

```python
folder_path = 'C:/Users/Enrico/Desktop/Datasets/Activity Logs'

dfs = []

for file_name in os.listdir(folder_path):
    if file_name.endswith(".csv"):
        file_path = os.path.join(folder_path, file_name)
        
        # Loading the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Adding the DataFrame to the list
        dfs.append(df)
```

```python
merged_df = pd.concat(dfs, ignore_index=True)
merged_df.info()
```

```python
activity_logs_df = merged_df
activity_logs_df.info()
```

```python
activity_logs_df['timestamp'] = pd.to_datetime(activity_logs_df["timestamp"])

# Estrazione della parte della data (anno-mese-giorno)
activity_logs_df['date'] = activity_logs_df['timestamp'].dt.strftime('%Y-%m-%d')
```

```python
activity_logs_df
```

```python
activity_logs_df_week_Nat = activity_logs_df[(activity_logs_df["date"] >= "2022-12-19") & (activity_logs_df["date"] <= "2022-12-25")  ]
```

```python
activity_logs_df_week = activity_logs_df[(activity_logs_df["date"] >= "2022-03-01") & (activity_logs_df["date"] <= "2022-03-06")  ]
```

```python
activity_logs_df_day = activity_logs_df_week[activity_logs_df_week["date"]== "2022-03-01"]
```
