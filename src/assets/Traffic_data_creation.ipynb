```python
import pandas as pd
from ast import literal_eval

# Leggi il dataset originale
df = pd.read_csv("C:/Users/Enrico/Desktop/Esame_Visual/public/Commute.csv")

# Inizializza un dizionario vuoto per tenere traccia delle occorrenze di currentLocations per ogni record
location_counts_per_record = {}

# Itera su ogni riga del dataset originale
for index, row in df.iterrows():
    # Estrai le currentLocations dalla riga corrente
    locations = literal_eval(row['currentLocations'])
    # Conta le occorrenze di ciascuna currentLocation nella riga corrente
    location_counts = dict(pd.Series(locations).value_counts())
    # Aggiungi il conteggio delle occorrenze al dizionario per il record corrente
    location_counts_per_record[index] = location_counts

# Crea un nuovo DataFrame per il nuovo dataset
new_df = pd.DataFrame(columns=['participantId', 'currentLocation', 'occorrenze'])

# Itera su ogni record e ogni currentLocation per costruire il nuovo DataFrame
for index, record in df.iterrows():
    # Estrai le occorrenze di currentLocation per il record corrente
    location_counts = location_counts_per_record[index]
    # Itera su ogni currentLocation e il relativo conteggio di occorrenze
    for location, count in location_counts.items():
        # Aggiungi una riga al nuovo DataFrame per ogni currentLocation
        new_df = new_df.append({
            'participantId': record['participantId'],
            'currentLocation': location,
            'occorrenze': count
        }, ignore_index=True)

# Salva il nuovo dataset in un nuovo file CSV
new_df
```

```python
new_df_sorted = new_df.sort_values(by='occorrenze', ascending=False)
```
